# Google Play Store Apps Dataset

This dataset contains information about applications available on the Google Play Store. It includes details such as app category, rating, number of reviews, size, installs, type (Free/Paid), price, content rating, genres, and the last update date.

The dataset can be used to analyze app performance, understand user preferences, and identify factors that influence app ratings and popularity.

## Business Problems

1. Developers may not know which factors contribute to higher app ratings.
2. Companies may want to understand what characteristics make an app more popular.
3. Businesses can use the analysis to improve app quality and increase user satisfaction.

## Machine Learning Problem

Type: Regression

Reason:
The target variable is the app Rating, which is a continuous numerical value. Therefore, predicting app ratings is a Regression problem.

## Target Variable

Rating
