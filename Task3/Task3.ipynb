# Key Insights

1. The dataset contains **4,009 used car records** with **12 features**, making it suitable for exploratory data analysis and future machine learning tasks.

2. The dataset consists of **11 categorical columns** and **1 numerical column (`model_year`)**, indicating that most features will require preprocessing and encoding before model training.

3. Missing values are present in several columns. The `fuel_type` column has **170 missing values**, the `accident` column has **113 missing values**, and the `clean_title` column has **596 missing values**, which need to be handled during data cleaning.

4. The vehicles in the dataset range from **1974 to 2024** model years. The median model year is **2017**, and the average model year is approximately **2015.5**, indicating that most vehicles are relatively recent.

5. The `milage` and `price` columns are stored as text (`object` data type) because they contain characters such as `"mi."` and `"$"`. These columns must be converted into numerical format before performing statistical analysis or building machine learning models.

```python
df.drop_duplicates(inplace=True)
```

```python
df['milage'] = df['milage'].str.replace(' mi.', '', regex=False)
df['milage'] = df['milage'].str.replace(',', '', regex=False)
df['milage'] = pd.to_numeric(df['milage'], errors='coerce')
```

```python
df['milage'] = df['milage'].fillna(df['milage'].median())
```

```python
df['price'] = df['price'].str.replace('$', '', regex=False)
df['price'] = df['price'].str.replace(',', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df['price'] = df['price'].fillna(df['price'].median())
```

```python
df['fuel_type'] = df['fuel_type'].fillna(df['fuel_type'].mode()[0])
df['accident'] = df['accident'].fillna(df['accident'].mode()[0])
df['clean_title'] = df['clean_title'].fillna(df['clean_title'].mode()[0])
```

```python
print("\nData Types After Cleaning:")
print(df.dtypes)
```

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
print("\nFirst 5 Rows of Cleaned Dataset:")
display(df.head())
```

```python
Q1 = df['price'].quantile(0.25)

Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR

upper = Q3 + 1.5*IQR

df = df[(df['price'] >= lower) & (df['price'] <= upper)]
```

```python
df['car_age'] = 2026 - df['model_year']
```

```python
df['mileage_per_year'] = df['milage'] / df['car_age']
```

```python
luxury = ['BMW','Mercedes-Benz','Audi']

df['is_luxury'] = df['brand'].isin(luxury).astype(int)
```

```python
df['price_category'] = pd.cut(
    df['price'],
    bins=3,
    labels=['Low','Medium','High']
)
```

```python
df['usage'] = df['milage'] / 1000
```

```python
df.to_csv("cleaned_used_cars.csv", index=False)
```
